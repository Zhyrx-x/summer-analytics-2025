### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
import pandas as pd

df = pd.read_csv('Cars.csv')

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df = df.dropna(subset=['horsepower', 'weight', 'mpg'])

median_mpg = df['mpg'].median()

above_median_cars = df[df['mpg'] > median_mpg].copy()

above_median_cars['hp_per_weight'] = above_median_cars['horsepower'] / above_median_cars['weight']

max_ratio = above_median_cars['hp_per_weight'].max()

best_ratio_cars = above_median_cars[above_median_cars['hp_per_weight'] == max_ratio][['name', 'horsepower', 'weight', 'mpg', 'hp_per_weight']]
print("Car(s) with the best horsepower-to-weight ratio among above-median mpg:")
print(best_ratio_cars)

```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('Cars.csv')


if 'model year' in df.columns:
    df = df.rename(columns={'model year': 'model_year'})
if 'year' in df.columns:
    df = df.rename(columns={'year': 'model_year'})


origin_map = {1: 'USA', 2: 'Europe', 3: 'Japan'}
if df['origin'].dtype != 'O':
    df['origin'] = df['origin'].map(origin_map)

avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()


plt.figure(figsize=(10,6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Evolution of Average MPG Over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.tight_layout()
plt.show()

```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('Cars.csv')


if 'model year' in df.columns:
    df = df.rename(columns={'model year': 'model_year'})
if 'year' in df.columns:
    df = df.rename(columns={'year': 'model_year'})


origin_map = {1: 'usa', 2: 'europe', 3: 'japan'}
if df['origin'].dtype != 'O':
    df['origin'] = df['origin'].map(origin_map)

plot_df = df.dropna(subset=['horsepower', 'weight', 'mpg', 'origin'])


plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=plot_df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200),
    alpha=0.7
)

plt.title('Horsepower vs Weight by Origin (Size = MPG)')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin and MPG', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()
plt.show()

```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

df = pd.read_csv('Cars.csv')

if 'model year' in df.columns:
    year_col = 'model year'
elif 'model_year' in df.columns:
    year_col = 'model_year'
elif 'year' in df.columns:
    year_col = 'year'
else:
    raise Exception("No year column found in the dataset.")


df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')
df = df.dropna(subset=['mpg'])


name_year_counts = df.groupby('name')[year_col].nunique()
multi_year_names = name_year_counts[name_year_counts > 1].index


stats = df[df['name'].isin(multi_year_names)].groupby('name').agg(
    num_years=(year_col, 'nunique'),
    count=('mpg', 'count'),
    mpg_std=('mpg', 'std'),
    avg_mpg=('mpg', 'mean')
).reset_index()


consistent_cars = stats[(stats['mpg_std'] < 1.0) & (stats['count'] >= 2)]


consistent_cars = consistent_cars.sort_values(['count', 'avg_mpg'], ascending=[False, False])

print(consistent_cars[['name', 'count', 'avg_mpg']])

```
